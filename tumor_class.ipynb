# Training

Due to the problem being a supervised classification problem, these will be the models we are going to compar to each other for now:
- Logistic Regression
- Super Vector Machine
- K-Nearest Neighbors
- Decison Tree
- Random Forest
- Naive Bayes
- XGBoost
- LightGBM

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
import xgboost as xgb
import lightgbm as lgb
```

```python
from sklearn.model_selection import train_test_split
```

For the metrics, we'll rely on:
- F1 (primarlly)
- Reacall
- Precision

```python
from sklearn.metrics import f1_score, recall_score, precision_score
```

## Preparing the Data

```python
X = df.drop(columns=['diagnosis'])
y = df['diagnosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Preparing the Models/ Score lists

```python
models = [LogisticRegression(random_state=42),
          KNeighborsClassifier(),
          SVC(probability=True,random_state=42),
          GaussianNB(),
          DecisionTreeClassifier(random_state=42),
          RandomForestClassifier(random_state=42),
          xgb.XGBClassifier(),
          lgb.LGBMClassifier(),]
model_names = ['LogisticRegression','KNN','SVM','NaiveBayes','DecisionTree','RandomForest','XGBoost','LightGBM']
```

```python
f1_scores = []
recall = []
precision = []
```

## Training the Models

```python
for model, name in zip(models, model_names):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    f1 = f1_score(y_test, y_pred)
    r = recall_score(y_test,y_pred)
    p = precision_score(y_test,y_pred)
    f1_scores.append(f1)
    recall.append(r)
    precision.append(p)
    print('%s: F1-score = %0.3f' % (name, f1))
    print('%s: Precision = %0.3f' % (name, p))
    print('%s: Recall = %0.3f' % (name, r))
    print('\n')
```

Out of the bunch, **LightGBM** preformed the best, we'll go with it.

## Another Try Using Cross-Validation/ GridSearch

```python
lgbm_class = lgb.LGBMClassifier(
    objective='binary',
    learning_rate=0.01,
    n_estimators=1000,  # Number of boosting rounds
    num_leaves=31,
    max_depth=-1,
    lambda_l1=1e-2,
    lambda_l2=1e-2,
    feature_fraction=0.8,
    bagging_fraction=0.8,
    bagging_freq=5
)
```

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold
```

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
f1_scores = cross_val_score(lgb_class, X, y, cv=cv, scoring='f1', n_jobs=-1)

print(f"F1 scores from each fold: {f1_scores}")
print(f"Mean F1 score: {f1_scores.mean()}")
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'num_leaves': [15, 31, 63],
    'learning_rate': [0.01, 0.05, 0.1],
    'n_estimators': [100, 500, 1000],
    'max_depth': [3, 5, 7]
}
```

```python
grid_search = GridSearchCV(
    estimator=lgbm_class,
    param_grid=param_grid,
    cv=cv,
    scoring='f1',
    n_jobs=-1
)
```

```python
grid_search.fit(X_train, y_train)
print(f"Best parameters: {grid_search.best_params_}")
print(f"Best cross-validated F1 score: {grid_search.best_score_}")
```
